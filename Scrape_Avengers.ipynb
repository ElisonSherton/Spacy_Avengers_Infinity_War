```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import spacy
%matplotlib inline
```

```python
import requests
r = requests.get('https://transcripts.fandom.com/wiki/Avengers:_Infinity_War')
print(len(r.text))
```

```python
soup = BeautifulSoup(r.text)
```

The attrs instance variable gives the attributes from a beautiful soup tag element. For example,

```python
soup.find_all('div')[2].attrs
```

All the dialogue are contained in the `div tag` whose `id attribute` is `mw-content-text`. This will help to extract all dialogue from the website.

All the characters' names are in bold, we can use this information to extract the names of authors.

```python
s = []
text = []
for link in soup.find_all('div'):
    # Everything is contained in the div tag whose id attribute is mw-content-text
    if ("id","mw-content-text") in link.attrs.items():
        # Every dialogue starts with the name of the speaker in bold
        for l in link.find_all('p'):
            if l.find("b"):
                # Extract only the author from the text
                author = re.findall(r'(^[A-Za-z].*?):', l.text)
                # Only keep the dialogues and remove all the actions mentioned in []
                text = re.sub(r'\[.*?\]',"",l.text)
                # Remove the speaker from the dialogue
                text = re.sub(r'.*?:',"", text)
                # Remove new line characters from the dialogue
                text = re.sub(r'\n','',text)
                # Remove the utf-8 Space
                # text = text.replace(u'\xa0', u' ')
                try:
                    s.append((author[0],text))
                except:
                    pass
```

Randomly read 10 dialogue from the transcript by taking them into a dataframe. 

```python
pd.set_option('display.max_colwidth',-1)
r = np.random.randint(100)

dialogue = pd.DataFrame(s, columns = ['Author', 'Text'])
dialogue[r:(r+10)]
```

# Tasks to accomplish:

- Overall top 10 verbs, nouns, adverbs and adjectives from the film.

- Top verbs and nouns spoke by a particular character.

- Top 30 named entities from the film.

- The similarity between the spoken lines of each character pair, e.g., the similarity between Thor’s and Thanos’ lines. 

```python
nlp = spacy.load('en_core_web_md')
```

```python
def top_10_pos(pos_match = "VERB"):
    # A dictionary to hold all the words
    parts_of_speech = {}
    
    # Checking if a token belongs to the part of speech specified as an input to the function.
    # Subsequently create a dictionary with these tokens as keys and their frequency as values
    
    for doc in nlp.pipe(dialogue.Text):
        for token in doc:
            if token.pos_ == pos_match:
                pos = token.lemma_.lower()
                if pos in parts_of_speech.keys():
                    parts_of_speech[pos] = parts_of_speech[pos] + 1
                else:
                    parts_of_speech[pos] = 1
                    
    # Sort the top 10 elements based on their occurence in the dialogues in a descending order.
    top_10_pos = sorted(parts_of_speech, key = lambda x: parts_of_speech[x], reverse = True)[:10]
    freq_top_10_pos = [parts_of_speech[i] for i in top_10_pos]
    
    # Plot the top 10 elements belonging to the pos_match part of speech
    rcParams['figure.figsize'] = 10,5
    sns.barplot(y = top_10_pos, x = freq_top_10_pos, orient = 'h')
    plt.xlabel('Top 10 {}'.format(pos_match.lower()))
    plt.ylabel('Frequency of {}'.format(pos_match.lower()))
    plt.show()   
```

# Characterwise Top 10 Parts of Speech

You can do a characterwise analysis. We can extract dialogues from a particular character and do the analysis which we did above. It can be done as follows.

```python
# Define the characters array containing which character's text you wanna analyze
characters = ['Steve Rogers','Tony Stark','Bruce Banner', 'Thanos', 'Thor', 'Ebony Maw']
dialogues = []
for i in characters:
    # Get all the dialogues of character i in a list
    dialogues = list(dialogue[dialogue.Author == i]['Text'])
    
    verbs = {}
    adjectives = {}
    nouns = {}
    adverbs = {}
    
    # The code below is similar to the top_10_pos function defined above. Just that we're
    # doing it 4 times for all 4 parts of speech (ie Nouns, Verbs, Adjectives, Adverbs) in one go.
    for doc in nlp.pipe(dialogues):
        for token in doc:
            if (token.pos_ == "VERB") and not (token.is_stop):
                pos = token.lemma_.lower()
                if pos in verbs.keys():
                    verbs[pos] = verbs[pos] + 1
                else:
                    verbs[pos] = 1

            elif token.pos_ == "NOUN" and not (token.is_stop):
                pos = token.lemma_.lower()
                if pos in nouns.keys():
                    nouns[pos] = nouns[pos] + 1
                else:
                    nouns[pos] = 1

            elif token.pos_ == "ADJ" and not (token.is_stop):
                pos = token.lemma_.lower()
                if pos in adjectives.keys():
                    adjectives[pos] = adjectives[pos] + 1
                else:
                    adjectives[pos] = 1

            elif token.pos_ == "ADV" and not (token.is_stop):
                pos = token.lemma_.lower()
                if pos in adverbs.keys():
                    adverbs[pos] = adverbs[pos] + 1
                else:
                    adverbs[pos] = 1
    
    top_10_nouns = sorted(nouns, key = lambda x: nouns[x], reverse = True)[:10]
    freq_top_10_nouns = [nouns[i] for i in top_10_nouns]
    
    fig, axes = plt.subplots(2, 2,figsize = (15,10))
    sns.barplot(x = top_10_nouns, y = freq_top_10_nouns, orient = "v", ax = axes[0,0])
    axes[0,0].title.set_text("Top 10 Nouns by {}".format(i))
    axes[0,0].set_xlabel("Top 10 Nouns.")
    axes[0,0].set_ylabel("Frequency Count of Top 10 Nouns.")
    axes[0,0].set_xticklabels(top_10_nouns, rotation=45)
    
    top_10_verbs = sorted(verbs, key = lambda x: verbs[x], reverse = True)[:10]
    freq_top_10_verbs = [verbs[i] for i in top_10_verbs]

    sns.barplot(x = top_10_verbs, y = freq_top_10_verbs, orient = "v", ax = axes[1,0])
    axes[1,0].title.set_text("Top 10 Verbs by {}".format(i))
    axes[1,0].set_xlabel("Top 10 Verbs.")
    axes[1,0].set_ylabel("Frequency Count of Top 10 Verbs.")
    axes[1,0].set_xticklabels(top_10_verbs, rotation=45)
    
    top_10_adj = sorted(adjectives, key = lambda x: adjectives[x], reverse = True)[:10]
    freq_top_10_adj = [adjectives[i] for i in top_10_adj]

    sns.barplot(x = top_10_adj, y = freq_top_10_adj, orient = "v", ax = axes[0,1])
    axes[0,1].title.set_text("Top 10 Adjectives by {}".format(i))
    axes[0,1].set_xlabel("Top 10 Adjectives.")
    axes[0,1].set_ylabel("Frequency Count of Top 10 Adjectives.")
    axes[0,1].set_xticklabels(top_10_adj, rotation=45)
    
    top_10_adv = sorted(adverbs, key = lambda x: adverbs[x], reverse = True)[:10]
    freq_top_10_adv = [adverbs[i] for i in top_10_adv]

    sns.barplot(x = top_10_adv, y = freq_top_10_adv, orient = "v", ax = axes[1,1])
    axes[1,1].title.set_text("Top 10 Adverbs by {}".format(i))
    axes[1,1].set_xlabel("Top 10 Adverbs.")
    axes[1,1].set_ylabel("Frequency Count of Top 10 Adverbs.")
    axes[1,1].set_xticklabels(top_10_adv, rotation=45)
    
    plt.tight_layout()
```

# Top 10 Named Entities

```python
dialogues = " ".join(list(dialogue.Text)).strip()
doc = nlp(dialogues)

named_ents = {}

for tk in doc.ents:
    token = tk.text
    if token in named_ents.keys():
        named_ents[token] = named_ents[token] + 1
    else:
        named_ents[token] = 1

top_10_named_ents = sorted(named_ents, key = lambda x:named_ents[x], reverse = True)[:10]
freq_top_10_ents = [named_ents[i] for i in top_10_named_ents]

rcParams['figure.figsize'] = 10,5
sns.barplot(x = top_10_named_ents, y = freq_top_10_ents, orient = 'v')
plt.xlabel('Top 10 Named Entities')
plt.ylabel('Frequency of Named Entities')
plt.show()   
```

# Character Similarity

```python
characters = ['Steve Rogers','Tony Stark','Bruce Banner', 'Thanos', 'Thor', 'Peter Parker', 'Ebony Maw']
dialogues = []
docs = {}

# Define a similarity matrix between the characters and set it to all 1s. Then we'lll systematically replace
# the non-diagonal elements with the similarity between the respective characters
similarities_matrix = np.full((len(characters), len(characters)), fill_value=1.0)

for i in characters:
    dialogues = list(dialogue[dialogue.Author == i]['Text'])
    corpus = []
    
    for doc in nlp.pipe(dialogues):
        for token in doc:
            if not(token.is_stop) and token.is_alpha:
                corpus.append(token.text)
    docs[i] = nlp(" ".join(corpus))
    
similarities = {}

for i in range(0,len(docs.items())):
    for j in range(i+1, len(docs.items())):
        similarities[characters[i] + " " + characters[j]] = docs[characters[i]].similarity(docs[characters[j]])
        similarities_matrix[i,j] = docs[characters[i]].similarity(docs[characters[j]])
        similarities_matrix[j,i] = docs[characters[j]].similarity(docs[characters[i]])
        
```

```python
for i in similarities.items():
    print("{0:<25}".format(i[0]), "{:0.2f}".format(i[1]))
```

```python
rcParams['figure.figsize'] = 10,7
sns.heatmap(similarities_matrix, yticklabels = characters, xticklabels = characters)
plt.xticks(rotation = 90)
plt.show()
```

# Latent Semantic Analysis

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from nltk.corpus import stopwords
SW = set(stopwords.words('english'))
```

```python
# Make a TFIDF Vector of the entire document corpus
# Documents are individual dialogues
# Terms are the text from those dialogues
# Remove punctuations, pronouns and stop words

doc = list(dialogue.Text)

Document_Corpus = []
for i in nlp.pipe(doc):
    docs = []
    for token in i:
        if token.is_alpha and not(token.is_stop) and not(token.lemma_ in SW) and not(token.lemma_ == "-PRON-"):
            # Remove additional punctuation if present in any word
            text = re.sub(r',\.:;!\?', '', token.lemma_)
            docs.append(text)
    Document_Corpus.append(" ".join(docs).strip())
        
tfidfVec = TfidfVectorizer().fit(Document_Corpus)
tfidfMat = tfidfVec.transform(Document_Corpus)
```

```python
# Number of topics to extract
# Let's select the top 10 topics
ntopics = 10
svd_model = TruncatedSVD(n_components = ntopics, algorithm = "randomized")
svd_model.fit(tfidfMat)
svd_model
```

```python
svd_model.components_.shape
```

```python
# Get the names of the features in order to figure out the important
# terms in the respective topics in subsequent SVD analysis
terms = tfidfVec.get_feature_names()
```

```python
# Get the weights of features(terms) in individual topics

for i, comp in enumerate(svd_model.components_):
    # Get a list of tuples with first entry as a term and second entry 
    # as it's weight in the ith topic
    term_component_tuple = zip(terms, comp)
    important_terms = sorted(term_component_tuple, key = lambda x:x[1], reverse = True)[:5]
    print("\nTopic {}: Top 5 important terms\n".format(str(i)))
    print("{0:<15}".format("Term"), "{0:<15}".format("Weight of Term"))
    for i in important_terms:
        print("{0:<15}".format(i[0]), "{:.3f}".format(i[1])) 
```

We can see that the topics don't give a whole lot of intuition about the latent topics. Let's try to see the explained variance of these top 10 topics.

```python
count = 1
for i in svd_model.explained_variance_:
    print("Topic {}".format(str(count)), "{:.4f}".format(i))
    count = count + 1
print("Cumulative Explained Variance: {:.4f}".format(np.sum(svd_model.explained_variance_)))
```

As we can see from the above distribution of topic importances, the first 10 topics only explain around 10% of the total variance in the data. Hence, there's no discerning meaning for the individual topics. Hence we get such seemingly gibberish top 5 words per topic.
